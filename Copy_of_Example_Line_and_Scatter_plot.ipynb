## Example with Finance data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as pdr
```

```python
stocks = ['GOOG', 'AMZN']
data = pdr.get_data_yahoo(stocks, start = '2021-01-01')['Close']
data.head()
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 8,6
plt.plot(data.AMZN, label = 'Amazon')
plt.plot(data.GOOG, label = 'Google')
plt.grid(True, color = 'k', linestyle = ':')
plt.title("Amazon and Google Stock prices")
plt.xlabel("Dates")
plt.ylabel("Prices")
plt.ylim(1500, 4000)
plt.style.use('fivethirtyeight')
plt.legend();
```

```python
plt.scatter(data.AMZN, data.GOOG)
```

```python
returns = np.log(data).diff()
returns.head()
returns.dropna(inplace = True)
```

```python
plt.scatter(returns.AMZN, returns.GOOG, alpha = 0.5)
```

```python
plt.axvline(0, c='gray', ls='--')
plt.axhline(0, c='gray', ls='--')
s = returns.AMZN
plt.scatter(returns.AMZN, returns.GOOG, c=returns.AMZN, cmap = 'nipy_spectral', s= s* 1000)
plt.colorbar(orientation = 'horizontal')
plt.ylim(-.05, 0.05)
plt.xlim(-.05, 0.05);
```
