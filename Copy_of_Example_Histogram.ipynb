```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896]
```

```python
plt.figure(figsize = (10,8))
plt.plot(years, apples, 's:b', ms = 8, mec='navy')
plt.plot(years, oranges, 'o--r', ms=10, mec = 'k')

plt.xlabel('Year')
plt.ylabel('Fruits')

plt.title('Fruits vs year')
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years, apples, linestyle='dashed', linewidth = 3,
         marker='o', markerfacecolor='blue', markersize=12)
plt.plot(years, oranges, linestyle='dashed', linewidth = 3,
         marker='o', markerfacecolor='blue', markersize=12)
plt.xlabel('Year')
plt.ylabel('Fruits')
plt.title('Fruits vs year')
plt.legend(['Apples', 'Oranges']);
```

# Histogram

```python
import numpy as np
import pandas as pd
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df.shape
```

```python
flowers_df.columns
```

```python
flowers_df.sepal_width.describe()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize = (8,6))
plt.hist(flowers_df.sepal_width, bins = [1,2,3,4,5,6]);
```

```python
flowers_df.species.unique()
```

```python
df1 = flowers_df[flowers_df.species == 'setosa']
df2 = flowers_df[flowers_df.species == 'versicolor']
df3 = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(df1.sepal_width, histtype = 'stepfilled', alpha = 0.5);
plt.hist(df2.sepal_width, histtype = 'stepfilled', alpha = 0.5);
plt.hist(df3.sepal_width, histtype = 'stepfilled', alpha = 0.5);
plt.legend(['setosa', 'versicolor', 'virginica']);
```

```python
plt.hist([df1.sepal_width, df2.sepal_width, df3.sepal_width], alpha = 0.7, color=['r', 'b', 'g'],
         bins = np.arange(2, 5, 0.5), stacked=True);
plt.legend(['setosa', 'versicolor', 'virginica']);
plt.title("Distribution of Sepal Width");
```
